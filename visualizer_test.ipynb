```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Each row: [x, y, q]
charges = np.array([
    [0.0, 0.0,  1e-6],   # 1 µC at origin
    [1.0, 0.0, -2e-6],   # -2 µC at (1, 0)
    [0.0, 1.0,  3e-6],   # 3 µC at (0, 1)
    [1.0, 1.0, -1e-6],   # -1 µC at (1, 1)
    [0.5, 0.5,  2e-6]    # 2 µC at (0.5, 0.5)
])

range_min =  np.min(charges[:,1]) - (np.mean(charges[:,1]))*0.25
range_max =  np.max(charges[:,1]) + (np.mean(charges[:,1]))*0.25
domain_min = np.min(charges[:,0]) - (np.mean(charges[:,0]))*0.25
domain_max = np.max(charges[:,0]) + (np.mean(charges[:,0]))*0.25 
x, y = np.meshgrid(np.linspace(domain_min, domain_max, num = 10), np.linspace(range_min, range_max, num = 10), indexing='xy')
x
```

```python
E_x_list = []
E_y_list = []

for charge in charges:
    r = x.copy() - charge[0]
    E_temp = (charge[2])/(r**2) * np.abs(r)/r #q/r^2*unit(r)
    E_x_list.append(E_temp)
    r = y.copy() - charge[1]
    E_temp = (charge[2])/(r**2) * np.abs(r)/r  #q/r^2*unit(r)
    E_y_list.append(E_temp)

#print(E_x_list)
#print(E_y_list)

E_x = E_x_list[0]
E_y = E_y_list[0]
for i in range(1, len(E_x_list)):
    E_x = E_x_list[i]
    E_y = E_y_list[i]

print(E_x)
print(E_y)
    
```

```python
print("x:", x)
print("E_x:", E_x)
print("y:", y)
print("E_y:", E_y)

plt.figure(figsize=(10, 10)) 
plt.streamplot(x,y,E_x,E_y, density=1.4, linewidth=None, color='#A23BEC')  
plt.title('Electromagnetic Field') 

```
